```python
import requests
import pandas as pd
import json
```

```python
#zebra
url = 'https://stockx.com/api/products/e90e1888-61f0-4681-8379-a4706e491235/activity'
params = {'state': 480,
         'currency': 'USD',
         'limit': 10,
         'page': 2,
         'sort': 'createdAt',
         'order': 'DESC',
         'country': 'US'}
header = {'User-Agent': 'Mozilla/5.0'}
x = requests.get(url, params = params, headers=header)
y=pd.DataFrame.from_records(json.loads(x.text)['ProductActivity'])
for i in range(50):
    params['page'] = i+2
    x = requests.get(url, params = params, headers=header)
    x = pd.DataFrame.from_records(json.loads(x.text)['ProductActivity'])
    y = y.append(x, ignore_index=True)
y
```
